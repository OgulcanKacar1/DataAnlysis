```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("calisan_data.csv")
df.head()
df.tail()
df.info()
df.describe()
```

```python
df.head()
```

```python
group__departman = df.groupby("Departman")["Isimler"].count()
group__departman
departmanList = df["Departman"].unique()
departmanList
```

```python
fig = plt.figure(figsize =(10,7))
group__departman.plot(kind="bar")
```

```python
#Departmandaki Calisanların Yüzdesini hesaplayalım
fig,ax = plt.subplots(figsize =(6,6))
ax.pie(group__departman,labels=departmanList,autopct='%1.1f%%')
plt.show()
```

```python
maas_ort = df.groupby("Departman")["Maas"].mean()
```

```python
#burada departmanlara göre maas ortalamamızı inceleyelim
fig,ax = plt.subplots(figsize =(8,5))
ax.plot(maas_ort,"red",marker=".",markersize =10,markerfacecolor="yellow")
plt.xticks(rotation = 90)
```

```python
#Aşağıdaki kodda dogum tarihi kolonundaki tarihlerin sadece yıl kısmını alıp buradaki değerleri int ' e dönüştürerek calısanlarımızın yas değerleriyle çalışabileceğiz.
```

```python
df["DogumTarihi"]=df["DogumTarihi"].str[-4:]
df["DogumTarihi"]= df["DogumTarihi"].astype(int)
df.dtypes
```

```python
calisan_Yas = 2023- df["DogumTarihi"]
print(max(calisan_Yas))
print(min(calisan_Yas)) #Burada en genç çalısanımızın 25 en yaslı calısanımızın 54 yasında oldugunu görüyoruz
```

```python
plt.hist(calisan_Yas)
plt.xlabel("Yaş")
```

```python
yas_Maas =df.groupby(pd.cut(2023-df['DogumTarihi'], bins=[20, 30, 40, 50,60]).cat.codes)["Maas"].mean()
calisan_yas_bolunmus = pd.cut(calisan_Yas, bins=[20,30,40,50,60])
```

```python
fig = plt.figure(figsize=(8,5))
plt.plot(yas_Maas)
```

```python
#Şimdi de çalışanlarımızın hangi telefon türlerini kullandıklarını inceleyelim
group_tel = df.groupby("TelefonTuru")["Isimler"].count()
group_tel
```
